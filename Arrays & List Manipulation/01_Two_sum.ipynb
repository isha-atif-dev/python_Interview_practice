#Q: Given an array of integers and a target value, return the indices of two numbers such that they add up to the target.

nums = [3, 2, 4, 7, 11, 15]

target = 9

We need to find two different indices whose values sum to the target. I’ll start with a straightforward brute-force approach.

```python
# Brute Force Solution
def two_sum(nums ,target):
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if nums[i] + nums[j] == target:
                return [i ,j]  
            
nums = [3, 2, 4, 7, 11, 15]
target = 9

two_sum(nums,target)
```

This works, but it has two nested loops, so it will check many pairs for large inputs.”

or 

"This works, but it’s inefficient for large inputs. We can optimise using a hash map to reduce lookup time."

complexity:


Outer loop runs n times

Inner loop runs about n times

So roughly n × n

So O(n²)

So we will go for the Optimized Solution as the brute force solution is not efficient

```python
def two_sum(nums,target):
    seen = {}

    for i,num in enumerate(nums):
        complement = target - num 

        if complement in seen:
            return [seen[complement],i]
        
        seen[num] = i

nums = [3, 2, 4, 7, 11, 15]
target = 9

two_sum(nums,target)
```

“What is O(n)? What does that even mean?”

Big O notation is a mathematical language used in computer science to describe the performance, scalability, and efficiency of an algorithm by measuring how its run time or space requirements grow as the input size (
) increases. 



Time complexity = how the number of operations grows as input size grows.
(Not actual seconds, but growth pattern.)

Space complexity = how much extra memory the algorithm uses as input size grows.
