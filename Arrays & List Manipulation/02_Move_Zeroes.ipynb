🎯 Exact Interview Question

Q: Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

You must do this in-place, meaning without making a copy of the array.

Input:

     nums = [0, 1, 0, 3, 12]


Output:

    [1, 3, 12, 0, 0]

Important:

-----Order of non-zero elements must stay the same.

------Modify the original list.

🔹 Step 1 — Brute Force Approach

1. Create a new list.

2. Add all non-zero numbers.

3. Count zeros.

4. Add zeros at the end.

5. Return new list.

```python
def move_zeroes(nums):
    non_zero = []
    zero_count = 0

    for num in nums:
        if num != 0:
            non_zero.append(num)
        else:
            zero_count += 1

    non_zero.extend([0]*zero_count)
    return non_zero

nums = [0, 1, 0, 3, 12]
move_zeroes(nums)
```

✅ Time Complexity

First loop:
      for num in nums:

This runs n times.

So first loop → O(n)


Second operation:
       non_zero.extend([0] * zero_count)


Creating [0] * zero_count
In worst case, zero_count could be n.

So creating that list takes → O(n)

Then extend() also takes O(n) in worst case.

So this part is also O(n).

Total Time

O(n) + O(n) = O(n)

We drop constants.

👉 Time Complexity = O(n)


✅ Space Complexity

We created:

   non_zero = []


In worst case, it can store all n elements.

We also created:

   [0] * zero_count


Which can also be size n in worst case.

So extra memory grows with input size.

👉 Space Complexity = O(n)

🎯 Move Zeroes — Optimized (In-Place)
